## 1.
#### How has honey production change from 1998 to 2021

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x="year",y="yieldpercol")
plt.title("Honey Production yield change from 1998 to 2021")
plt.xlabel("Year 1998 to 2021")
plt.ylabel("Yield per colony ")
plt.grid(True)
plt.show()
```

## Observation:
#### The Yield per colony in the US has been decreasing over the years.

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x="year",y="totalprod")
plt.title("Total honey Production yield change from 1998 to 2021")
plt.xlabel("Year 1998 to 2021")
plt.ylabel("Total Honey Production ")
plt.grid(True)
plt.show()
```

## Observation:

#### The overall honey production in the US has been decreasing over the years.

## 2.

#### Over time, what are the major production trends across the states?

```python
plt.figure(figsize=(24,8))
sns.barplot(x=df["State"],y=df["totalprod"], ci=None, hue=df["year"])
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,3))
g=sns.FacetGrid(df,col="State", col_wrap=4, size=3)
g=g.map(plt.plot,"year","totalprod")
plt.show()
```

## Observation
#### Statewise Total Honey Production. North Dakota,California, South Dakota, Florida, Montana are the top Producer.

## 3.
#### Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006, which was when concern over Colony Collapse Disorder spread nationwide?

```python
# Collecting the data of the number of honey producing colonies and yield per colonies
```

```python
df_before_2006=df[df["year"]<2006]
```

```python
df_before_2006.head()
```

```python
df_before_2006.shape
```

```python
state_colonies=df_before_2006.groupby("State")["numcol"].sum()
state_yield=df_before_2006.groupby("State")["yieldpercol"].sum()
```

```python
state_colonies.head(10)
```

```python
state_yield.head()
```

```python
plt.figure(figsize=(25,25))
plt.subplot(2,1,1)
plt.plot(state_colonies.index,state_colonies.values)
plt.title("Number of Honey Producing Colonies (pre-2006)")
plt.xticks(rotation=90)
plt.subplot(2,1,2)
plt.plot(state_yield.index,state_yield.values)
plt.title("Yield per Colony (pre-2006)")
plt.xticks(rotation=90)
plt.show()
```

## Observation :
1. State wise Number of Honey Producing Colonies (Pre-2006)<br>
2. State wise Yield per Colony (pre-2006)


## 4.

#### Are there any patterns that can be observed between total honey production and value of production every year ?

```python
# Create a scatter plot to visualize the relationalship between honey production
```

```python
plt.scatter(df["totalprod"], df["prodvalue"])
plt.xlabel("Total honey production")
plt.ylabel("Value of production ")
plt.show()
```

## Observation:

Total honey production and value of production every year are increasing.


## 5. 


How has the value of production , which in some sense could be tied to demand, changed every year?

```python
# Group the data by year and calculate the total value of production for each year
value_by_year = df.groupby("year")["prodvalue"].sum()
```

```python
plt.plot(value_by_year)
plt.xlabel("Year")
plt.ylabel("Value of production")
plt.grid(True)
plt.show()
```

## Observation:

As the overall Honey production is decreasing, the total value of production for each year is increasing.
